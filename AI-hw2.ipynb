```python
import random

class hill_climbing:
    def __init__(self, cuts, length, max_iter):
#         initialization
        self.init = cuts
        self.len = length       
        self.maxim = max_iter
        self.iteration = 0
        self.ans = cuts
        self.best_score =self.hill_climber(self.init)
    
    def hill_climber(self, answer):
        current = compute(answer, self.len)
        self.iteration += 1
#         break if max iteration is met
        if (self.iteration == self.maxim):
            self.ans = answer
            return current
#         generate the first neighbor which has the lower value
        best_neighbor = generate_neighbors(answer, self.len, current)
        if(compute(best_neighbor,self.len) < current):
            return self.hill_climber(best_neighbor)
        else:
#             if none were found we are in  local optimum
            self.ans = answer
            return current

;
# find the first neighbor which lower value
# neighbors are permutations resulted from swaping two indexes
def generate_neighbors(permutation, length, value):
    n = len(permutation)
    for i in range(n - 1):
        for j in range(i + 1, n):
            neighbor = list(permutation)
            neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
            if(compute(neighbor,length) < value):
                return neighbor

    return permutation

# Compute the number of rolls needed with given pattern of cuts
def compute(pattern, length):
    answer = 0
    temp = length
    for i in range(len(pattern)):
        if(temp - pattern[i] >= 0):
            temp -= pattern[i]
        elif(i != len(pattern) - 1):
            temp = length - pattern[i]
            answer += 1
        if(i == len(pattern) - 1):
            answer += 1
    return answer


    
        
        
    
```

```python
# Function for running algorithm multiple times

def hill(cuts, length, aim, max_iter,iteration):
    example = hill_climbing(cuts, length, max_iter)
    best = example.best_score
    ans = example.ans
    for i in range(iteration):
        random.shuffle(cuts)
        example = hill_climbing(cuts, length, max_iter)
        if best > example.best_score:
            best = example.best_score
            ans = example.ans
    
    return (best,ans)
        
    
```

```python
import random
import math

class SA:
    def __init__(self, cuts, length, temp = 10000, alpha = 0.98, T_min = 0.001):
#         initialization
        self.init = cuts
        self.len = length       
        self.t_min = T_min
        self.t = temp
        self.alpha = alpha
        self.ans = cuts
        self.best_score =self.iterate(self.init)

    
    def iterate(self, answer):
        current = compute(answer, self.len)
#         iterate until t is still smaller than the declared minimum
        if (self.t < self.t_min):
            self.ans = answer
            return current
#         generate a random neighbor by swaping two random indexes
        neighbor = generate_neighbor_SA(answer)
        if(compute(neighbor,self.len) < current):
#             if it has better value change the current answer to it
            self.t = self.t*self.alpha
            return self.iterate(neighbor)
        
        else:
#             if not accept it with the given probability
            r = random.uniform(0, 1)
            p = math.exp((current - compute(neighbor,self.len))/self.t)
            self.t = self.t*self.alpha
            if (r < p):
                return self.iterate(neighbor)
            else:
                return self.iterate(answer)

;

def generate_neighbor_SA(permutation):
    n = len(permutation)
    i = random.randint(0, n - 1)
    j = i
    
    while j == i:
        j = random.randint(0, n - 1)
    
    neighbor = list(permutation)
    neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
    
    return neighbor
    

def compute(pattern, length):
    answer = 0
    temp = length
    for i in range(len(pattern)):
        if(temp - pattern[i] >= 0):
            temp -= pattern[i]
        elif(i != len(pattern) - 1):
            temp = length - pattern[i]
            answer += 1
        if(i == len(pattern) - 1):
            answer += 1
    return answer
```

```python
# Function for running algorithm multiple times
def simulatedA(cuts, length):
    example = SA(cuts, length)
    best = example.best_score
    ans = example.ans
    for i in range(100):
        random.shuffle(cuts)
        example = SA(cuts, length)
        if best > example.best_score:
            best = example.best_score
            ans = example.ans
    return (best,ans)
```

```python
import random

class genetic:
    
    def __init__(self, cuts, length, alpha, iteration = 1000):
#         initialization
        self.leng = length
        self.alpha = alpha
        self.iteration = 0
        self.max_iter = iteration
        self.ans = cuts
        self.population = self.build_init(cuts)
        self.score = self.iterate()
        
        
    
    def build_init(self, cuts):
#         build up your initial population with random permutations of input
        population = []
        population.append(cuts)
        for i in range(50):
            temporary = cuts.copy()
            random.shuffle(temporary)
            population.append(temporary)
#        sort it based on the fittness function
        population_sorted = sorted(population, key=lambda x: compute(x, self.leng))
        return population_sorted
    
    def iterate(self):
#         for each iteration
        for i in range(self.max_iter):
#           compute kids
            kids = self.generate_kids(self.population)
            for kid in kids:
#             mutate each child with the probability of alpha
                r = random.uniform(0, 1)
                if(self.alpha > r):
                    kid = self.mutation(kid)
#                 Replace the kids in population
            self.update(kids)
#            Sort population again
            self.population = sorted(self.population, key=lambda x: compute(x, self.leng))
#       answer is the best member of the population
        self.ans = self.population[0]
        return compute(self.population[0], self.leng)
    
#  generate kids by randomly choosing two parrents from population and then performing crossover
    def generate_kids(self, parents):                
        n = len(parents)
        kid = []
        for _ in range(20):
            i = random.randint(0, n - 1)
            j = i

            while j == i:
                j = random.randint(0, n - 1)
            kid.append(crossover([parents[i], parents[j]]))
            return kid
#  perform mutation by swaping two random indexes    
    def mutation(self, kid):
        neighbor = list(kid)
        n = len(kid)
        i = random.randint(0, n - 1)
        j = i

        while j == i:
            j = random.randint(0, n - 1)

        neighbor[i], neighbor[j] = neighbor[j], neighbor[i]

        return neighbor

#    put the child in the population only if its performance is better than the weakest member of the population    
    def update(self, kids):
        for kid in kids:
            if (compute(kid, self.leng) < compute(self.population[-1], self.leng)):
                self.population[-1] = kid
;

def compute(pattern, length):
    answer = 0
    temp = length
    for i in range(len(pattern)):
        if(temp - pattern[i] >= 0):
            temp -= pattern[i]
        elif(i != len(pattern) - 1):
            temp = length - pattern[i]
            answer += 1
        if(i == len(pattern) - 1):
            answer += 1
    return answer

# perform cross over by randomly chopsing a point
def crossover(parents):
    lenn = random.randint(0, len(parents[0]) - 1)
    child = [parents[0][i] for i in range(lenn)]
    parent1_copy = parents[1].copy()
    for i in range(lenn):
        for j in range(len(parent1_copy) - 1, 0, -1):
            if(parents[0][i] == parent1_copy[j]):
                parent1_copy[j] = None
                break
   
    for i in range(len(parent1_copy)):
        if(parent1_copy[i] != None):
            child.append(parent1_copy[i])
    return child
```

```python
def gen(cuts, length, alpha, iteration = 100):
    example = genetic(cuts, length, alpha)
    best = example.score
    ans = example.ans
    for i in range(iteration):
        example = genetic(cuts, length, alpha)
        if best > example.score:
            best = example.score
            ans = example.ans
    return (best,ans)
```
